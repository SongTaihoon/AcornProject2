```python
df = pd.read_csv('./food.csv', sep=',', encoding='cp949')
okt=Okt()
ii=[]
AA=[]
BB=[]
CC=[]
DD=[]
EE=[]
FF=[]
AA_1=[]

for i in range(len(df)):
    AA.append(df.iloc[i,4]) #리뷰
    AA_1.append(df.iloc[i,4]) #리뷰원본
    BB.append(df.iloc[i,2]) #리뷰어
    CC.append(df.iloc[i,3]) #제목
    DD.append(df.iloc[i,0]) #가게이름
    EE.append(df.iloc[i,5]) #개인평점
    FF.append(df.iloc[i,9]) #권역
    
for i in range(len(AA)):
    AA[i]=re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣]","",AA[i])
    AA[i]=okt.morphs(AA[i],norm=True,stem=True)
    AA[i]=[x for x in AA[i] if len(x)>1]
    AA[i]=' '.join(AA[i])
    
df1=pd.DataFrame({'review':AA,'ID':BB,'title':CC,'place':DD,'rating':EE,'region':FF,'raw':AA_1})

hangul = re.compile('[ㄱ-ㅣ가-힣]+')

for i in range(len(df1)):
    if len(hangul.findall(df1.iloc[i,0]))>1:
        if len(df1.iloc[i,0])>1:
            ii.append(i)
```

```python
df2=df1.iloc[ii,:]
df2.to_csv('foodnoun.csv',index=False, encoding='cp949')
df2.head()
```

```python
#단어 빈도 추출 - 카테고리 생성(가격/맛/서비스/웨이팅?)
A=df2.review.to_list()
B=[]
for i in range(len(A)):
    B+=A[i].split(' ')
    
words_count={}
for word in B:
    if word in words_count:
        words_count[word] += 1
    else:
        words_count[word] = 1
sorted_words = sorted([(k,v) for k,v in words_count.items()], key=lambda word_count: -word_count[1])
print([w[0]for w in sorted_words])
```

```python
#Word2Vec 카테고리별 코사인 유사도가 높은 단어 구하기
X=[[]]*len(df2)
for i in range(len(df2)):
    X[i] = df2.iloc[i,0].split(' ')

from gensim.models import Word2Vec
model_w = Word2Vec(sentences=X,vector_size=500,window=5,min_count=2,workers=4,sg=1)
```

```python
noum1=pd.DataFrame(model_w.wv.most_similar('맛있다',topn=20),columns=['단어','맛있다'])
noum2=pd.DataFrame(model_w.wv.most_similar('가격',topn=20),columns=['단어','가격'])
noum3=pd.DataFrame(model_w.wv.most_similar('위치',topn=20),columns=['단어','위치'])
noum4=pd.DataFrame(model_w.wv.most_similar('서비스',topn=20),columns=['단어','서비스'])
noum5=pd.DataFrame(model_w.wv.most_similar('분위기',topn=20),columns=['단어','분위기'])
```

```python
A=list(set(noum1.단어.tolist()+noum2.단어.tolist()+noum3.단어.tolist()+noum4.단어.tolist()+noum5.단어.tolist()))
A=sorted(A)
pd.DataFrame(A).to_csv('noumlist.csv',index=False, encoding='utf-8-sig')
```

```python
#카테고리-단어 코사인유사도행렬 구하기
A=pd.read_csv('./noumlist.csv',sep=',',encoding='utf-8-sig')
A=A.iloc[:,0].tolist()
A=sorted(A)

B_맛있다=[]
B_가격=[]
B_위치=[]
B_서비스=[]
B_분위기=[]
for i in range(len(A)):
    B_맛있다.append(model_w.wv.similarity('맛있다',A[i]))
    B_가격.append(model_w.wv.similarity('가격',A[i]))
    B_위치.append(model_w.wv.similarity('위치',A[i]))
    B_서비스.append(model_w.wv.similarity('서비스',A[i]))
    B_분위기.append(model_w.wv.similarity('분위기',A[i]))
B_total = pd.DataFrame({'맛있다':B_맛있다,'가격':B_가격,'위치':B_위치,'서비스':B_서비스,'분위기':B_분위기}).transpose()
B_total.columns=[A]
B_total[B_total<0.8]=0
B_total.to_csv('noumlist2.csv',index=False, encoding='utf-8-sig')
```
