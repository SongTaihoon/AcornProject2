```python
def find_sim_res(df, sorted_ind, reviews, top_n=10):
    #인자로 입력된 reviews DF에서 'review_content' 칼럼이 입력된 reviews 값인 DF추출
    review_content = df[df['review_content'] == reviews]
    review_index = review_content.index.values
    similar_indexes = sorted_ind[review_index,:(top_n)]
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)
    return df.iloc[similar_indexes]
```
