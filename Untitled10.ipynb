```python
df = pd.read_csv('./food.csv', encoding='cp949')
df
```

```python
def text_clearing(text): 
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
    # 지정한 정규식에 해당하지 않은 것은 길이가 0인 문자열로 변환한다.
    result = hangul.sub('', text)
    return result
```

```python
df["ko_review_title"] = df["review_title"].apply(lambda x : text_clearing(x))
df["ko_review_content"] = df["review_content"].apply(lambda x : text_clearing(x))
```

```python
df
```

```python
df.drop(["review_title","review_content"], axis=1, inplace=True)
```

```python
# konlpy 라이브러리로 텍스트 데이터에서 형태소를 추출한다.
def get_pos (x) :
    tagger = Okt()
    pos = tagger.pos(x)
    
    # 단어와 품사를 합쳐서 하나의 단어로 만들어준다.
    result = []
    
    # 형태소의 수만큼 반복한다.
    # 조사인 것과 명사인 것이 같을 수 있기 때문에 구분해준다.
    # 형태소 벡터를 만들때 추후 사용
    for a1 in pos :
        result.append(f'{a1[0]}/{a1[1]}')
    
    return result
```

```python
df["content"]=df["ko_review_title"]+""+df["ko_review_content"]
```

```python
#평점 전처리
def star_preprocessing(value) :
    if value <=35 :
        return '0'
    else :
        return '1'
    
#형태소 분석
def tokenizer(text) :
    okt = Okt()
    return okt.morphs(text)
```

```python
def step1_data_preprocessing() :
    df1 = pd.read_csv('./hotel.csv', encoding='cp949')
    
    df1['user_rating'] = df1['user_rating'].apply(star_preprocessing)
    
    text_list = df1['review_content'].tolist()
    star_list = df1['user_rating'].tolist()
    
    text_train, text_test, star_train, star_test = train_test_split(text_list, star_list, test_size=0.2, random_state = 0)
    
    return text_train, text_test, star_train, star_test
    
```

```python
def step2_learning(X_train, X_test, y_train,  y_test):
    tfidf = TfidfVectorizer(lowercase=False, tokenizer=tokenizer)
    
    logistic = LogisticRegression(C=10.0, penalty ='l2', random_state = 0)
    
    pipe = Pipeline([('vect', tfidf),('clf', logistic)])
    
    pipe.fit(X_train, y_train)
    
    y_pred = pipe.predict(X_test)
    print(accuracy_score(y_test, y_pred))
    
    with open('hotelpipe.dat', 'wb') as fp :
        pickle.dump(pipe, fp)
        
    print('저장완료')
```

```python
def step3_using_model() :
    with open('hotelpipe.dat','rb') as fp :
        pipe = pickle.load(fp)
        
    while True :
        text = input('리뷰를 작성해주세요 :')
        
        str = [text]
        r1 = np.max(pipe.predict_proba(str) * 100)
        
        r2 = pipe.predict(str)[0]
        
        if r2 == '1' :
            print('긍정적인 리뷰')
        else :
            print('부정적인 리뷰')
            
        print('정확도 : %.3f' % r1)
        
def learning() :
    text_train, text_test, star_train, star_test = step1_data_preprocessing()
    step2_learning(text_train, text_test, star_train, star_test)
    
def using() :
    step3_using_model()
```
